```python
#1.Replace the null values

data=data.fillna(0)
```

```python
#2. Deal with outliers
columns_to_filter = ['people_vaccinated', 'daily_vaccinations_raw', 'total_vaccinations_per_hundred', 'people_vaccinated_per_hundred']

# Calculate the mean for each of the specified columns
means = data[columns_to_filter].mean()

# Filter the DataFrame based on the mean of each specified column
for col in columns_to_filter:
    data = data[(data[col] >= means[col]) & (data[col] <= means[col])]
```

```python
#Replace categorical columns with dummy data
cc=data[['country','iso_code','date','vaccines','source_name','source_website']]
data=pd.get_dummies(data,columns=['country','iso_code','date','vaccines','source_name','source_website'])
```
